````python
os.chdir("/Users/mariapalafox/Desktop/ensembl-vep/CpDCysLys_mapping")
vep = "VEP_detected_filtered_missenseonly_header_removed.txt"
vep = pd.read_csv(vep, delimiter="\t", low_memory=False)
````

````python
disease = vep[['#Uploaded_variation', 'Location', 'Allele', 'Gene', 'Feature',
       'Feature_type', 'Consequence', 'cDNA_position', 'CDS_position',
       'Protein_position', 'Amino_acids', 'Codons', 'SYMBOL', 'Existing_variation',
       'IMPACT','GENE_PHENO','MAX_AF', 'CLIN_SIG', 'SOMATIC', 'PHENO']].copy()
````

````python
# filtering out '-' rows in 'CLIN_SIG'
clinvar = disease[disease['CLIN_SIG'] != '-'].copy()
clinvar
````

````python
# dropping Feature column then deleting all duplicate rows
clinvar.drop(['Feature','cDNA_position','CDS_position','Protein_position'], axis=1, inplace=True)

clinvar.drop_duplicates(inplace=True)
clinvar.reset_index(inplace=True)
clinvar
````
