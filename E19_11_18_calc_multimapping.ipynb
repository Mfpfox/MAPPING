## create new columns for version number

````python
def split_version(df):
    pro = df["ENSPv"].str.split(".", n=1, expand=True)
    pro.columns = ['stable', 'pro_ver']
    ensp = pro['pro_ver'].astype(int)
    
    tx = df["ENSTv"].str.split(".", n=1, expand=True)
    tx.columns = ['stable', 'tx_ver']
    enst = tx['tx_ver'].astype(int)
    
    gen = df["ENSGv"].str.split(".", n=1, expand=True)
    gen.columns = ['stable', 'gen_ver']
    ensg = gen['gen_ver'].astype(int)
    
    final = pd.concat([df, ensp, enst, ensg], axis=1)
    return final
````

````python
v85 = split_version(v85)
v92 = split_version(v92)
v94 = split_version(v94)
v96 = split_version(v96)
v97 = split_version(v97)

# same for shared ENSG_ENST_ENSP version
shared85 = split_version(shared85)
shared92 = split_version(shared92)
shared94 = split_version(shared94)
shared96 = split_version(shared96)
shared97 = split_version(shared97)
````

## create count frame from counting len of set's in each column

````python
%%time

def setme(s):
    ls = list(s)
    finalset = set(ls)
    return finalset


# columns based on set of IDs or sequences from all releases
ENSP_all_3963 = all_3963.groupby('xref')['ENSP'].apply(setme)
ENSPv_all_3963 = all_3963.groupby('xref')['ENSPv'].apply(setme)
ENST_all_3963 = all_3963.groupby('xref')['ENST'].apply(setme)
ENSTv_all_3963 = all_3963.groupby('xref')['ENSTv'].apply(setme)
ENSG_all_3963 = all_3963.groupby('xref')['ENSG'].apply(setme)
ENSGv_all_3963 = all_3963.groupby('xref')['ENSGv'].apply(setme)
stableID_key_all_3963 = all_3963.groupby('xref')['stableID_key'].apply(setme)
proSequence_all_3963 = all_3963.groupby('xref')['proSequence'].apply(setme)

# concat
set_columns = pd.concat([ENSP_all_3963, ENSPv_all_3963, ENST_all_3963, ENSTv_all_3963, ENSG_all_3963, ENSGv_all_3963, stableID_key_all_3963, proSequence_all_3963], axis=1)
````

````python
set_columns['count_ENSP'] = set_columns['ENSP'].apply(lambda x: len(x))
set_columns['count_ENSPv'] = set_columns['ENSPv'].apply(lambda x: len(x))
set_columns['count_ENST'] = set_columns['ENST'].apply(lambda x: len(x))
set_columns['count_ENSTv'] = set_columns['ENSTv'].apply(lambda x: len(x))
set_columns['count_ENSG'] = set_columns['ENSG'].apply(lambda x: len(x))
set_columns['count_ENSGv'] = set_columns['ENSGv'].apply(lambda x: len(x))
set_columns['count_stableID_key'] = set_columns['stableID_key'].apply(lambda x: len(x))
set_columns['count_proSequence'] = set_columns['proSequence'].apply(lambda x: len(x))
````
