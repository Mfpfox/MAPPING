def split_version(df):
    pro = df["ENSPv"].str.split(".", n=1, expand=True)
    pro.columns = ['stable', 'pro_ver']
    ensp = pro['pro_ver'].astype(int)
    
    tx = df["ENSTv"].str.split(".", n=1, expand=True)
    tx.columns = ['stable', 'tx_ver']
    enst = tx['tx_ver'].astype(int)
    
    gen = df["ENSGv"].str.split(".", n=1, expand=True)
    gen.columns = ['stable', 'gen_ver']
    ensg = gen['gen_ver'].astype(int)
    
    final = pd.concat([df, ensp, enst, ensg], axis=1)
    return final

````python
%%time

# diff number of rows but same number of IDs in each release

def setme(s):
    ls = list(s)
    finalset = set(ls)
    return finalset


# columns based on set of IDs or sequences from all releases
ENSP_all_3953 = all_3953.groupby('xref')['ENSP'].apply(setme)
ENSPv_all_3953 = all_3953.groupby('xref')['ENSPv'].apply(setme)
ENST_all_3953 = all_3953.groupby('xref')['ENST'].apply(setme)
ENSTv_all_3953 = all_3953.groupby('xref')['ENSTv'].apply(setme)
ENSG_all_3953 = all_3953.groupby('xref')['ENSG'].apply(setme)
ENSGv_all_3953 = all_3953.groupby('xref')['ENSGv'].apply(setme)
stableID_key_all_3953 = all_3953.groupby('xref')['stableID_key'].apply(setme)
proSequence_all_3953 = all_3953.groupby('xref')['proSequence'].apply(setme)

# concat
set_columns = pd.concat([ENSP_all_3953, ENSPv_all_3953, ENST_all_3953, ENSTv_all_3953, ENSG_all_3953, ENSGv_all_3953, stableID_key_all_3953, proSequence_all_3953], axis=1)

set_columns['count_ENSP'] = set_columns['ENSP'].apply(lambda x: len(x))
set_columns['count_ENSPv'] = set_columns['ENSPv'].apply(lambda x: len(x))
set_columns['count_ENST'] = set_columns['ENST'].apply(lambda x: len(x))
set_columns['count_ENSTv'] = set_columns['ENSTv'].apply(lambda x: len(x))
set_columns['count_ENSG'] = set_columns['ENSG'].apply(lambda x: len(x))
set_columns['count_ENSGv'] = set_columns['ENSGv'].apply(lambda x: len(x))
set_columns['count_stableID_key'] = set_columns['stableID_key'].apply(lambda x: len(x))
set_columns['count_proSequence'] = set_columns['proSequence'].apply(lambda x: len(x))
````
