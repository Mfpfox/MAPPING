## making dictionary with key as 'ref' and value as 'cadd_phred' | 'GERP++' | 'M-CAP score'

````python
# g6 and cadd
g6cadd = {}
for i in g6sort['pymol_ref'].unique(): 
    g6cadd[i] = [g6sort['CADD_phred'][j] for j in g6sort[g6sort['pymol_ref']==i].index]
g6cadd
````

## averaging missense CADD phred scores (v1.4 for GRCh38) over each amino acid in CASP8 and G6PD

````python
avgC8 = {}
for k,v in c8cadd.items():
    avgC8[k] = sum(v)/float(len(v))
avgC8
````

````python
maxG6 = {}
for k,v in g6cadd.items():
    maxG6[k] = max(v)
maxG6
````

````python
avgG6 = {}
for k,v in g6cadd.items():
    avgG6[k] = sum(v)/float(len(v))
avgG6
````

````python
G6score = pd.DataFrame.from_dict(avgG6,orient='index')
G6max = pd.DataFrame.from_dict(maxG6,orient='index')
G6score = G6score.reset_index() #changing index into a column
G6max = G6max.reset_index() #changing index into a column
G6score.columns =['pymol_reference', 'avgMissense_CADDv1.4_GRCh38']
G6max.columns =['pymol_reference', 'avgMissense_CADDv1.4_GRCh38']
````

````python
# G6PD
# split ref column into aa and position
splitscore = G6score["pymol_reference"].str.split("_",n=1, expand=True)
splitscore.columns = ["originalAmino", "protein_pos"]
finalG6 = pd.concat([G6score, splitscore],axis=1)

# max
# split ref column into aa and position
splitscore = G6max["pymol_reference"].str.split("_",n=1, expand=True)
splitscore.columns = ["originalAmino", "protein_pos"]
finalG6max = pd.concat([G6max, splitscore],axis=1)
````

````python
# CASP8
C8score = pd.DataFrame.from_dict(avgC8,orient='index')
C8score = C8score.reset_index() #changing index into a column
C8score.columns =['pymol_reference', 'avgMissense_CADDv1.4_GRCh38']
# split ref column into aa and position
splitscore = C8score["pymol_reference"].str.split("_",n=1, expand=True)
splitscore.columns = ["originalAmino", "protein_pos"]
finalC8 = pd.concat([C8score, splitscore],axis=1)
````
