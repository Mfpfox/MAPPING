```python
# spliting dataframe and adding columns for assembly version of cadd score then recombine files
def assemblyGrouper(df, CK):
    if CK == 'C':       
        test2 = df[['pos_id19', 'pos_id38', 'xref', 'matched_target', 'lost_amino', 'gained_amino',
       'Amino_acids', 'pos_ID', 'Cys_reactivity', 'Cys_react_threshold',
       'Cys_target_label', 'geneNamePrimary']].copy()
        test3 = test2.copy()
        test2['CADD_score'] = df['CADD_phred_hg19'].copy()
        test2['assembly'] = 'hg19'
        test3['CADD_score'] = df['CADD_phred_hg38'].copy()
        test3['assembly'] = 'hg38'
        test4 = pd.concat([test2,test3])
        test4.reset_index(inplace=True,drop=True)
        print("combined shape : ", test4.shape)
        print("copy shapes : ", test2.shape, test3.shape)
        return test4
        
    if CK == 'K':
        test2 = df[['pos_id19', 'pos_id38', 'xref', 'matched_target', 'lost_amino', 'gained_amino',
       'Amino_acids', 'pos_ID', 'Lys_reactivity', 'Lys_react_threshold','Lys_target_label','geneNamePrimary']].copy()
        test3 = test2.copy()
        test2['CADD_score'] = df['CADD_phred_hg19'].copy()
        test2['assembly'] = 'hg19'
        test3['CADD_score'] = df['CADD_phred_hg38'].copy()
        test3['assembly'] = 'hg38'   
        test4 = pd.concat([test2,test3])
        test4.reset_index(inplace=True,drop=True)
        print("combined shape : ", test4.shape)
        print("copy shapes : ", test2.shape, test3.shape)
        return test4
    
    if CK == 'N':
        test2 = df[['pos_id19', 'pos_id38', 'xref', 'matched_target','lost_amino', 'gained_amino',
       'Amino_acids', 'pos_ID_falseCKtarget', 'geneNamePrimary']].copy()
        test3 = test2.copy()
        test2['CADD_score'] = df['CADD_phred_hg19'].copy()
        test2['assembly'] = 'hg19'
        test3['CADD_score'] = df['CADD_phred_hg38'].copy()
        test3['assembly'] = 'hg38'   
        test4 = pd.concat([test2,test3])
        test4.reset_index(inplace=True,drop=True)
        print("combined shape : ", test4.shape)
        print("copy shapes : ", test2.shape, test3.shape)
        return test4
```

```python
# spliting dataframe and adding columns for assembly version of cadd score then recombine files
def assemblyGrouper(df, CK):
    if CK == 'C':       
        test2 = df[['index', 'xref','Cys_reactivity', 'Cys_react_threshold', 'Cys_target_label','geneNamePrimary']].copy()
        test3 = test2.copy()
        test2['CADD_avg'] = df['hg19_pos_avg'].copy()
        test2['CADD_max'] = df['hg19_pos_max'].copy()
        test2['assembly'] = 'hg19'
        test3['CADD_avg'] = df['hg38_pos_avg'].copy()
        test3['CADD_max'] = df['hg38_pos_max'].copy()
        test3['assembly'] = 'hg38'
        test4 = pd.concat([test2,test3])
        test4.reset_index(inplace=True,drop=True)
        print("combined shape : ", test4.shape)
        print("copy shapes : ", test2.shape, test3.shape)
        return test4
        
    if CK == 'K':
        test2 = df[['index','xref','Lys_reactivity', 'Lys_react_threshold', 'Lys_target_label','geneNamePrimary']].copy()
        test3 = test2.copy()
        test2['CADD_avg'] = df['hg19_pos_avg'].copy()
        test2['CADD_max'] = df['hg19_pos_max'].copy()
        test2['assembly'] = 'hg19'
        test3['CADD_avg'] = df['hg38_pos_avg'].copy()
        test3['CADD_max'] = df['hg38_pos_max'].copy()
        test3['assembly'] = 'hg38'
        test4 = pd.concat([test2,test3])
        test4.reset_index(inplace=True,drop=True)
        print("combined shape : ", test4.shape)
        print("copy shapes : ", test2.shape, test3.shape)
        return test4
    
    if CK == 'N':
        test2 = df[['index','xref','geneNamePrimary']].copy()
        test3 = test2.copy()
        test2['CADD_avg'] = df['hg19_pos_avg'].copy()
        test2['CADD_max'] = df['hg19_pos_max'].copy()
        test2['assembly'] = 'hg19'
        test3['CADD_avg'] = df['hg38_pos_avg'].copy()
        test3['CADD_max'] = df['hg38_pos_max'].copy()
        test3['assembly'] = 'hg38'
        test4 = pd.concat([test2,test3])
        test4.reset_index(inplace=True,drop=True)
        print("combined shape : ", test4.shape)
        print("copy shapes : ", test2.shape, test3.shape)
        return test4
```
