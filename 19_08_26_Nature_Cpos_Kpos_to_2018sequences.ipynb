````python
# finding overlap between cys and lysine reactivity with 2018 positions

# making list of react pos_ID column
CreactList = list(Creact.pos_ID)
KreactList = list(Kreact.pos_ID)
print(len(CreactList))
print(len(KreactList))

C2018 = list(Cys2018)
K2018 = list(Lys2018) 
````

````python
# CYS reactivity overlap with CYS2018
overlapCysreact = set(C2018).intersection(CreactList)
overlapLysreact = set(K2018).intersection(KreactList)
print(len(overlapCysreact))
print(len(overlapLysreact))
````

````python
# label the original data frame with new column whether its in CCDS 2018 or not
def addcolumnconditional(mapList, df, dfcol, newcol): 
    mendel = []
    for g in df[dfcol]:
        if g in mapList: 
            mendel.append("True")
        else:
            mendel.append("False")
    df.loc[:,newcol] = mendel
    return df
CreactOverlapped = addcolumnconditional(overlapCysreact, Creact, "pos_ID", "2018pos_ID")
KreactOverlapped = addcolumnconditional(overlapLysreact, Kreact, "pos_ID", "2018pos_ID")
````

````python
# seperating True and False then saving False
CreactTRUEonly = CreactOverlapped[CreactOverlapped["2018pos_ID"] == "True"]
CreactFALSEonly = CreactOverlapped[CreactOverlapped["2018pos_ID"] == "False"]
print(CreactTRUEonly.shape)
print(CreactFALSEonly.shape)
````

````python
# seperating True and False then saving False
KreactTRUEonly = KreactOverlapped[KreactOverlapped["2018pos_ID"] == "True"]
KreactFALSEonly = KreactOverlapped[KreactOverlapped["2018pos_ID"] == "False"]
print(KreactTRUEonly.shape)
print(KreactFALSEonly.shape)
````

````python
# saving rows not found in 2018 file
CreactFALSEonly.to_csv("Cys_reactivity_notFoundin2018positions_49pos.csv",index=False)
KreactFALSEonly.to_csv("Lys_reactivity_notFoundin2018positions_75pos.csv",index=False)
````
