## LIST and MISCAST file cleaned above ready to be merged with AA-level dbnsfp and CADD files

````python
# simplify cadd file to score columns
simcol = ['#Chrom', 'Pos', 'Ref', 'Alt','GeneName','RawScore', 'PHRED']
cadd37 = cadd37[simcol].copy()
cadd37.columns= ['#Chrom', 'Pos', 'Ref', 'Alt','GeneName','CADD37_raw', 'CADD37_phred']
cadd37['dna_key'] = cadd37['#Chrom'].astype(str) + \
'_' + cadd37['Pos'].astype(str) + \
'_' + cadd37['Ref'].astype(str) + \
'_' + cadd37['Alt'].astype(str)
cadd37 = cadd37[['dna_key','#Chrom', 'Pos', 'Ref', 'Alt','GeneName','CADD37_raw', 'CADD37_phred']]
cadd37.head(3)
````

````python
simcol = ['hg19_chr', 'hg19_pos(1-based)', 'ref', 'alt', 'aaref','aaalt','CADD_raw', 'CADD_phred','matched_UKBID','matched_aapos', 'matched_index']
dbnsfp = dbnsfp[simcol].copy()
dbnsfp.columns = ['hg19_chr', 'hg19_pos', 'ref', 'alt','aaref','aaalt', 'CADD38_raw', 'CADD38_phred','matched_UKBID','matched_aapos', 'matched_index']
dbnsfp['dna_key'] = dbnsfp['hg19_chr'].astype(str) + \
'_' + dbnsfp['hg19_pos'].astype(str) + \
'_' + dbnsfp['ref'].astype(str) + \
'_' + dbnsfp['alt'].astype(str)
dbnsfp = dbnsfp[['dna_key','hg19_chr', 'hg19_pos', 'ref', 'alt','aaref','aaalt', 'CADD38_raw', 'CADD38_phred','matched_UKBID','matched_aapos', 'matched_index']]
dbnsfp.head(3)
````

````python
## merging cadd and dbnsfp file on dna_key
merged = pd.merge(cadd37,dbnsfp, how='inner',on=['dna_key'])
````

````python
# remove weird values from merge of dbnsfp and cadd for 1 protein
merged = merged[merged['matched_aapos'] != -1]
merged = merged[merged['aaalt'] != 'X']
merged = merged[merged['aaalt'] != '.']
merged = merged[merged['aaref'] != 'X']
merged = merged[merged['aaref'] != '.']
````

````python
## check canonical seq. length
print(len(merged.matched_aapos.unique()))
## making new column for missense type & sorting df by 'matched_aapos'
merged['missense_type'] = merged['aaref'] + '/' + merged['aaalt']
# sort protein positions
merged = merged.sort_values(by=['matched_aapos'])
merged.reset_index(drop=True, inplace=True)
````

````python
merged['pymol_ref'] = merged['aaref'] + '_' + merged['matched_aapos'].astype(str)
````

````python
merged.head(10)
# this would be good point to save file for specific missense change level analysis!!!!
````

````python
# group by pymol ref M_1
list37 = {}
list38 = {}
for i in merged['pymol_ref'].unique(): 
    list37[i] = [merged['CADD37_phred'][j] for j in merged[merged['pymol_ref']==i].index]
    list38[i] = [merged['CADD38_phred'][j] for j in merged[merged['pymol_ref']==i].index]

## average and max missense scores for each pymol_ref
avg37 = {}
for k,v in list37.items():
    avg37[k] = sum(v)/float(len(v))

max37 = {}
for k,v in list37.items():
    max37[k] = max(v)
    
avg38 = {}
for k,v in list38.items():
    avg38[k] = sum(v)/float(len(v))

max38 = {}
for k,v in list38.items():
    max38[k] = max(v)
    
## making into dataframe and adding to simplified original file
caverage37 = pd.DataFrame.from_dict(avg37,orient='index')
caverage37 = caverage37.reset_index() 
cmax37 = pd.DataFrame.from_dict(max37,orient='index')
cmax37 = cmax37.reset_index() 
caverage37.columns = ['pymol_reference', 'avg_CADD37']
cmax37.columns = ['pymol_reference', 'max_CADD37']

caverage38 = pd.DataFrame.from_dict(avg38,orient='index')
caverage38 = caverage38.reset_index() 
cmax38 = pd.DataFrame.from_dict(max38,orient='index')
cmax38 = cmax38.reset_index() 
caverage38.columns = ['pymol_reference', 'avg_CADD38']
cmax38.columns = ['pymol_reference', 'max_CADD38']

# split ref column into aa and position
csplitavg37 = caverage37["pymol_reference"].str.split("_",n=1, expand=True)
splitcol = ["AAref", "Position"]
csplitavg37.columns = splitcol

finalavg = pd.concat([csplitavg37, caverage37,caverage38, cmax37,cmax38],axis=1)
#finalavg=finalavg[['pymol_reference', 'avg_CADD37','avg_CADD38', 'max_CADD37','max_CADD38']].copy()
````
